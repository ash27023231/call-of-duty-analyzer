# Rigs Vs Modes

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


plt.rcParams['figure.figsize'] = [15, 10]
# data to plot
n_groups = 3

# these were calculated using MS excel's filter
FTL = (252, 186, 158) 
Merc = (252, 185, 145)    
Phantom = (252, 164, 80)
Stryker = (252, 185, 94)
Warfighter = (0, 24, 155)


# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.15
opacity = 0.8

rects1 = plt.bar(index, FTL, bar_width,
                 alpha=opacity,
                 color='black',
                 label='FTL')
 
rects2 = plt.bar(index + bar_width, Merc, bar_width,
                 alpha=opacity,
                 color='green',
                 label='Merc')

rects3 = plt.bar(index + 2*bar_width, Phantom, bar_width,
                 alpha=opacity,
                 color='blue',
                 label='Phantom')

rects4 = plt.bar(index + 3*bar_width, Stryker, bar_width,
                 alpha=opacity,
                 color='red',
                 label='Stryker')

rects5 = plt.bar(index + 4*bar_width, Warfighter, bar_width,
                 alpha=opacity,
                 color='yellow',
                 label='Warfighter')

 
plt.xlabel('Modes')
plt.ylabel('Rigs Frequency')
plt.title('Rigs Vs Modes')
plt.xticks((np.arange(0, 2, step=0.9)+ 0.4) , ('Hardpoint', 'Search&Destroy', 'Uplink'))
plt.legend()
 
plt.tight_layout()
plt.show()
```

# Weapons Vs Maps

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


plt.rcParams['figure.figsize'] = [15, 10]
# data to plot
n_groups = 7

# these were calculated using MS excel's filter
NV4 = (232, 186, 0, 171, 3, 70, 46) 
KBAR32 = (215, 243, 62, 167, 110, 83, 66)      
KBSLongbow = (0, 7, 0, 6, 0, 0, 0)         
Erad = (17, 155, 176, 104, 111, 87, 64)         
Karma45 = (0, 1, 2, 0, 0, 0, 0)


# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.19
opacity = 0.8

rects1 = plt.bar(index, NV4, bar_width,
                 alpha=opacity,
                 color='b',
                 label='NV4')
 
rects2 = plt.bar(index + bar_width, KBAR32, bar_width,
                 alpha=opacity,
                 color='g',
                 label='KBAR-32')

rects3 = plt.bar(index + 2*bar_width, KBSLongbow, bar_width,
                 alpha=opacity,
                 color='r',
                 label='KBSLongbow')

rects4 = plt.bar(index + 3*bar_width, Erad, bar_width,
                 alpha=opacity,
                 color='y',
                 label='Erad')
rects5 = plt.bar(index + 4*bar_width, Karma45, bar_width,
                 alpha=opacity,
                 color='black',
                 label='Karma-45')
 
plt.xlabel('Maps')
plt.ylabel('Weapon Frequency')
plt.title('Weapons Vs Maps')
plt.xticks((np.arange(0, 7, step=1) + 0.4) , ('Retaliation', 'Throwback', 'Frost', 'Breakout', 'Scorch', 'Crusher', 'Precinct' ))
plt.legend()
 
plt.tight_layout()
plt.show()

```

# Weapons Accuracy

```python

import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


plt.rcParams['figure.figsize'] = [15, 10] 
objects = ('NV4', 'KBAR-32', 'KBSLongbow', 'Erad', 'Karma-45')
y_pos = np.arange(len(objects))

# these were calculated using MS excel's filter/function
Average_Accuracy = [0.212525424, 0.215565539, 0.261307692, 0.221441176, 0.191333333]
 
plt.barh(y_pos, Average_Accuracy, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Accuracy')
plt.title('Weapons Accuracy')
 
plt.show()
```
